## **Determinar o p-valor**

O p-valor √© uma medida utilizada em testes de hip√≥teses estat√≠sticas para ajudar a determinar a signific√¢ncia dos resultados obtidos. Ele representa a probabilidade de se obter um resultado t√£o extremo (ou mais extremo) quanto o observado na amostra, assumindo que a hip√≥tese nula (H‚ÇÄ) seja verdadeira. Portanto, o valor-p √© um indicador que nos ajuda a entender se a hip√≥tese nula pode ou n√£o ser rejeitada com base nos dados coletados.

**Passos para Compara√ß√£o do p-valor com Œ±**

Comparar o p-valor com o n√≠vel de signific√¢ncia (ùõº) √© um passo crucial no processo de teste de hip√≥teses estat√≠sticas.

Vamos detalhar esse processo:


**1. Definir o N√≠vel de Signific√¢ncia (Œ±):**


  * O n√≠vel de signific√¢ncia (ùõº) √© o limite m√°ximo de probabilidade de cometer um erro do tipo I (rejeitar a hip√≥tese nula quando ela √© verdadeira) que a pessoa pesquisadora est√° disposto a aceitar. Comumente, valores de ùõº s√£o 0.01, 0.05, ou 0.10.

  * Por exemplo, Œ± = 0.05 significa que a pessoa pesquisadora est√° disposto a aceitar uma chance de 5% de cometer um erro do tipo I.

**2. Calcular o p-valor:**

* O p-valor √© a probabilidade de observar uma estat√≠stica de teste t√£o extrema quanto (ou mais extrema que) a observada, assumindo que a hip√≥tese nula √© verdadeira.
* O c√°lculo do p-valor depende da distribui√ß√£o de probabilidade espec√≠fica usada no teste (normal, t-student, qui-quadrado, etc.).

**3. Comparar o p-valor com Œ±:**

* Se ùëù ‚â§ ùõº:
        * Rejeitamos a hip√≥tese nula (H‚ÇÄ).                
        * A evid√™ncia contra H‚ÇÄ √© forte o suficiente para concluir que √© improv√°vel que o resultado observado tenha ocorrido por acaso sob H‚ÇÄ.

* Se ùëù > ùõº:
        * N√£o rejeitamos a hip√≥tese nula (H‚ÇÄ).
        * N√£o h√° evid√™ncias suficientes para concluir que o resultado observado √© inconsistente com H‚ÇÄ.

### **Tipos de Testes de Hip√≥teses**

**Testes Param√©tricos:** Esses testes s√£o feitos presumindo que os dados seguem uma distribui√ß√£o normal. **Testes N√£o Param√©tricos:** S√£o testes estat√≠sticos especiais que n√£o precisam considerar que os dados seguem uma distribui√ß√£o normal. Isso os torna √∫teis e adapt√°veis quando n√£o podemos garantir essa suposi√ß√£o.

* Da mesma forma, a √°rea √† direita de qualquer valor positivo de z representa a probabilidade acumulada de obter um valor maior que aquele z-score.

**Exemplo Pr√°tico**

### Exemplo de teste de hip√≥teses:

A hip√≥tese nula (H‚ÇÄ) √© a afirma√ß√£o de que a turma tenha a nota m√©dia de 70.

A hip√≥tese alternativa (H‚ÇÅ) √© a afirma√ß√£o de que a turma tenha a nota m√©dia diferente de 70.

```python
from statsmodels.stats.weightstats import ztest
import numpy as np

# Dados da amostra
amostra = np.array([68, 70, 72, 65, 75, 78, 67, 69, 74, 71, 73, 72, 68, 70, 71])
valor_referencia = 70

print(f"Amostra {amostra}")
print(f"Valor Refer√™ncia: {valor_referencia}")
```

```python
# Teste z para uma amostra
z_statistic, p_value = ztest(amostra, value=valor_referencia)
print("Teste z para uma amostra:")
print(f"z-statistic: {z_statistic:.4f}, p-value: {p_value:.4f}")
```

**Teste z para uma amostra:**

*ztest(amostra, value=valor_referencia)* compara a m√©dia da amostra com o valor_referencia.

*z_statistic* √© o valor da estat√≠stica z calculada.

*p_value* √© o valor p associado ao teste, que nos diz se a diferen√ßa observada √© estatisticamente significativa.

```python
# Dados das amostras independentes
amostra1 = np.array([68, 70, 72, 65, 75, 78, 67, 69, 74, 71])
amostra2 = np.array([60, 62, 64, 63, 66, 70, 65, 67, 71, 69])

# Teste z para duas amostras independentes
z_statistic_ind, p_value_ind = ztest(amostra1, amostra2)
print("\nTeste z para duas amostras independentes:")
print(f"z-statistic: {z_statistic_ind:.4f}, p-value: {p_value_ind:.4f}")
```

**Teste z para duas amostras independentes:**

*ztest(amostra1, amostra2)* compara as m√©dias de amostra1 e amostra2.

*z_statistic_ind* √© o valor da estat√≠stica z calculada.

*p_value_ind* √© o valor p associado ao teste, que indica se as diferen√ßas entre as duas amostras s√£o estatisticamente significativas.

Para realizar um **teste t de Student** usando Python, voc√™ pode utilizar a biblioteca **scipy**, que cont√©m fun√ß√µes para estat√≠sticas.

Aqui est√° um exemplo de c√≥digo que realiza tanto um **teste t** para uma amostra **√∫nica** quanto um **teste t** para duas amostras **independentes**:


**Teste t para uma amostra**

**Teste t para duas amostras independentes**

**Vamos implementar esses dois testes.**

Suponha que temos uma amostra de pontua√ß√µes de exame de estudantes e queremos verificar se a m√©dia da amostra √© significativamente diferente de um valor de refer√™ncia (por exemplo, 70). Tamb√©m vamos comparar duas amostras independentes de pontua√ß√µes de exame de dois grupos diferentes de estudantes.

```python
import numpy as np
from scipy import stats

# Dados da amostra
amostra_notas = [68, 70, 72, 65, 75, 78, 67, 69, 74, 71, 73, 72, 68, 70, 71]
valor_referencia = 70

# Teste t para uma amostra
t_statistic, p_value = stats.ttest_1samp(amostra_notas, valor_referencia)
print("Teste t para uma amostra:")
print(f"t-statistic: {t_statistic:.4f}, p-value: {p_value:.4f}")
```

Este teste verifica se a m√©dia da amostra √© significativamente diferente de um valor de refer√™ncia (70 neste caso).

**Teste t para uma amostra:**

*stats.ttest_1samp(amostra, valor_referencia)* compara a m√©dia da amostra com o valor_referencia.

*t_statistic* √© o valor da estat√≠stica *t* calculada.

## Cen√°rio do teste de hip√≥teses:

Queremos testar se um novo m√©todo de ensino melhora o desempenho dos estudantes em compara√ß√£o com o m√©todo tradicional j√° utilizado.

**Passo a passo:**

1. Definir as hip√≥teses:

    **Hip√≥tese nula (H0‚Äã):** O novo m√©todo de ensino n√£o tem efeito (a m√©dia das notas dos estudantes com o novo m√©todo √© igual √† m√©dia das notas dos estudantes com o m√©todo tradicional).

    **Hip√≥tese alternativa (H1‚Äã):** O novo m√©todo de ensino tem efeito (a m√©dia das notas dos estudantes com o novo m√©todo √© diferente da m√©dia das notas dos estudantes com o m√©todo tradicional).

2. Coletar dados:

    **Grupo 1 (m√©todo tradicional):** uma amostra de 40 estudantes.
    
    **Grupo 2 (novo m√©todo):** uma amostra de 40 estudantes.

3. Calcular a m√©dia e o desvio padr√£o dos dois grupos.

4. Calcular o Z score e o p-valor.

5. Tomar uma decis√£o com base no p-valor e no n√≠vel de signific√¢ncia (Œ±).

```python
# bibliotecas necess√°rias:
import numpy as np
from statsmodels.stats.weightstats import ztest
```

```python
# Gerar dados simulados
np.random.seed(42)  # Para reprodutibilidade
```

```python
# Amostras do m√©todo tradicional (grupo controle)
tradicional = np.random.normal(loc=70, scale=12, size=40)  # M√©dia 70, desvio padr√£o 12
tradicional
```

```python
# Amostras do novo m√©todo (grupo teste)
novo_metodo = np.random.normal(loc=75, scale=12, size=40)  # M√©dia 75, desvio padr√£o 12
novo_metodo
```

```python
# Calcular m√©dias e desvios padr√£o
mean_tradicional = np.mean(tradicional)
mean_novo_metodo = np.mean(novo_metodo)
std_tradicional = np.std(tradicional, ddof=1)
std_novo_metodo = np.std(novo_metodo, ddof=1)
```

```python
# Realizar o teste z
z_score, p_value = ztest(novo_metodo, tradicional)

# N√≠vel de signific√¢ncia
alpha = 0.05

print(f"M√©dia do grupo tradicional: {mean_tradicional:.2f}")
print(f"M√©dia do grupo novo m√©todo: {mean_novo_metodo:.2f}")
print(f"Desvio padr√£o do grupo tradicional: {std_tradicional:.2f}")
print(f"Desvio padr√£o do grupo novo m√©todo: {std_novo_metodo:.2f}")
print(f"Z score: {z_score:.2f}")
print(f"P-valor: {p_value:.4f}")

if p_value < alpha:
    print("Rejeitamos a hip√≥tese nula. H√° evid√™ncias de que o novo m√©todo de ensino tem efeito.")
else:
    print("N√£o rejeitamos a hip√≥tese nula. N√£o h√° evid√™ncias de que o novo m√©todo de ensino tem efeito.")
```
